```python
# Importation des bibliothèques et modules nécessaires au fonctionnement de ce notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Visualisation de `WindGustSpeed`

```python
# Vérification de l'existence de stations météo ayant zéro entrée dans la colonne « WindGustSpeed »

df_location_windgustspeed_counts = df[["Location", "WindGustSpeed"]].groupby("Location").count()
df_location_windgustspeed_counts[df_location_windgustspeed_counts["WindGustSpeed"] == 0]
```

```python
# Création d'un graphique en boîtes à moustaches pour « WindGustSpeed »

## Initialisation de la toile
fig, ax = plt.subplots(figsize = (12, 8))

## Traçage
sns.boxplot(x = df["Location"].sort_values(ascending = True), 
            y = df["WindGustSpeed"], 
            color = "#588157", 
            flierprops = {"marker": "x", 
                          "markeredgecolor": "#a3b18a"})

## Mise en forme
plt.xticks(rotation = 90)
plt.ylim(0, 140)
plt.ylabel("WindGustSpeed (km / h)")
plt.title("Vitesse maximale des rafales de vent par station météo", fontweight = "bold");

### Mise en évidence des stations météo ne disposant pas de données pour « WindGustSpeed »
for index in range(len(df_location_windgustspeed_counts)):
    if df_location_windgustspeed_counts.iloc[index][0] == 0:
        ax.get_xticklabels()[index].set_color("#dda15e")
        ax.get_xticklabels()[index].set_weight("bold")
```

```python
# Test d'inversion des axes

## Initialisation de la toile
fig, ax = plt.subplots(figsize = (12, 16))

## Traçage
sns.boxplot(y = df["Location"].sort_values(ascending = True), 
            x = df["WindGustSpeed"], 
            color = "#588157", 
            flierprops = {"marker": "x", 
                          "markeredgecolor": "#a3b18a"})

## Mise en forme
plt.xlim(0, 140)
plt.xlabel("WindGustSpeed (km / h)", fontweight = "bold")
plt.ylabel("Location", fontweight = "bold")
plt.title("Vitesse maximale des rafales de vent par station météo", fontweight = "bold", fontsize = 16);

### Mise en évidence des stations météo ne disposant pas de données pour « WindGustSpeed »
for index in range(len(df_location_windgustspeed_counts)):
    if df_location_windgustspeed_counts.iloc[index][0] == 0:
        ax.get_yticklabels()[index].set_color("#dda15e")
        ax.get_yticklabels()[index].set_weight("bold")
```

## 5. Création d'une rose des vents : exemple des rafales de vent à Albury

```python
# Création d'un nouveau DataFrame « df_wind » contenant uniquement l'emplacement des stations météo (« Location ») ainsi que les variables relatives au vent

df_wind = df[["Location", "WindGustDir", "WindGustSpeed", "WindDir9am", "WindSpeed9am", "WindDir3pm", "WindSpeed3pm"]]
df_wind.head()
```

```python
# Suppression des lignes contenant au moins un NaN

df_wind = df_wind.dropna()
df_wind.isna().sum()
```

```python
# Création d'un nouveau DataFrame « df_wind_gust » par filtrage de « df_wind » afin d'étudier spécifiquement les rafales de vent

df_wind_gust = df_wind[["Location", "WindGustDir", "WindGustSpeed"]]
df_wind_gust.head()
```

```python
# Création d'un nouveau DataFrame « df_wind_gust_albury » par filtrage de « df_wind_gust » afin d'étudier spécifiquement les rafales de vent à l'échelle de la station météo d'Albury

df_wind_gust_albury = df_wind_gust[df_wind_gust["Location"] == "Albury"]
df_wind_gust_albury.head()
```

```python
# Examen de l'ordre de fréquence des modalités de « WindGustDir »

df_wind_gust_albury["WindGustDir"].value_counts()
```

```python
# Création d'un dictionnaire pour convertir, dans « WindGustDir », les directions du vent en radians

WindGustDir_directions_to_radians_dictionary = {
    "N": np.pi / 2,
    "E": 0.0,
    "S": 3 * np.pi,
    "W": np.pi,
    "NE": np.pi / 4,
    "SE": 7 * np.pi / 4,
    "SW": 5 * np.pi / 4,
    "NW": 3 * np.pi / 4,
    "NNE": 0.375 * np.pi,
    "ENE": 0.125 * np.pi,
    "ESE": 1.875 * np.pi,
    "SSE": 1.625 * np.pi,
    "SSW": 1.375 * np.pi,
    "WSW": 1.125 * np.pi,
    "WNW": 0.875 * np.pi,
    "NNW": 0.625 * np.pi
}
```

```python
# Exécution de la conversion

df_wind_gust_albury = df_wind_gust_albury.replace(to_replace = WindGustDir_directions_to_radians_dictionary)
df_wind_gust_albury.head()
```

```python
# Découpage de « WindGustSpeed » en 6 catégories

wind_gust_speed_categories = ["0 - 20", "20 - 40", "40 - 60", "60 - 80", "80 - 100", "> 100"]
wind_gust_speed_cut = pd.cut(x = df_wind_gust_albury["WindGustSpeed"], 
                             bins = (0, 20, 40, 60, 80, 100, np.inf), 
                             labels = wind_gust_speed_categories)
```

```python
# Création d'un tableau croisé pour le traçage du graphique

wind_gust_table = pd.crosstab((df_wind_gust_albury["WindGustDir"]), wind_gust_speed_cut)
wind_gust_table
```

```python
# Initialisation des paramètres du graphique

## Définition du nombre de tranches à 16, le nombre de directions représentées par une rose des vents
N = 16

## Génération de 16 angles « theta », chacun de valeur 2 * pi / 16 radians, pour chacune des tranches
theta = np.linspace(0.0, 2 * np.pi, N, endpoint = False)

## Création d'une variable « width » correspondant à la largeur des tranches
width = np.pi / 8

## Initialisation de l'origine
bottom = 0.0

## Définition de la palette de couleurs
colors = ["#caf0f8", "#90e0ef", "#00b4d8", "#0077b6", "#03045e", "#000000"]
```

```python
# Création de la rose des vents pour les plus fortes rafales de vent à Albury

## Initialisation de la toile
fig = plt.figure(figsize = (6 ,6))

## Traçage
ax = plt.subplot(projection = "polar")
for i in range(wind_gust_table.shape[1]): ### Itération sur les 6 colonnes (catégories de vent) de « wind_gust_table »
    radii = wind_gust_table.iloc[:, i] ### Colonne de « wind_gust_table » avec 16 lignes, chacune correspondant à la fréquence d'une direction du vent
    ax.bar(x = theta, 
           height = radii, 
           width = width, 
           bottom = bottom, 
           label = wind_gust_speed_categories[i] + " km / h", 
           color = colors[i], 
           alpha = 0.75) ### Traçage de la colonne
    bottom += radii ### Empilement de la colonne suivante

## Mise en forme
lines, labels = plt.thetagrids(np.arange(0.0, 360, 22.5), 
                               ("E", "ENE", "NE", "NNE", 
                                "N", "NNW", "NW", "WNW", 
                                "W", "WSW", "SW", "SSW", 
                                "S", "SSE", "SE", "ESE"))
ax.set_yticklabels([])
plt.title("Les plus fortes rafales de vent à Albury", fontweight = "bold", fontsize = 16)
plt.legend(bbox_to_anchor=(1.25, 0), fontsize = 10);
```
